Cleaning Data:
There are some columns which are in numeric form but it include the Commas instead of dots.
so as we have to use these values as an integer.
So Firstly we will replace the ","(Commas) with "."(Dots)
And then convert each of the column to numeric form.
we don't need these columns as a String Type.
That's why we will convert it to Numeric form.

```python
df['Happiness Score']=df['Happiness Score'].astype(str).str.replace(',','.')
df['Economy (GDP per Capita)']=df['Economy (GDP per Capita)'].astype(str).str.replace(',','.')
df['Family (Social Support)']=df['Family (Social Support)'].astype(str).str.replace(',','.')
df['Health (Life Expectancy)']=df['Health (Life Expectancy)'].astype(str).str.replace(',','.')
df['Freedom']=df['Freedom'].astype(str).str.replace(',','.')
df['Trust (Government Corruption)']=df['Trust (Government Corruption)'].astype(str).str.replace(',','.')
df['Generosity']=df['Generosity'].astype(str).str.replace(',','.')
df["Happiness Score"] = pd.to_numeric(df["Happiness Score"])
df["Economy (GDP per Capita)"] = pd.to_numeric(df["Economy (GDP per Capita)"])
df["Family (Social Support)"] = pd.to_numeric(df["Family (Social Support)"])
df["Health (Life Expectancy)"] = pd.to_numeric(df["Health (Life Expectancy)"])
df["Freedom"] = pd.to_numeric(df["Freedom"])
df["Trust (Government Corruption)"] = pd.to_numeric(df["Trust (Government Corruption)"])
df["Generosity"] = pd.to_numeric(df["Generosity"])
```
